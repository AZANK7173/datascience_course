- Supondo que descobrimos que todos os valores da coluna 1 estão incorretos e que deveríamos somar 1 unidade a cada um deles. Utilizar **df.apply** para adicionar essa quantidade aos valores.
- Mudar todos os valores da coluna price maiores do que 50000 para NaN. 
Usar **value_counts** para contar os valores da coluna 1 do dataset.

### Bônus

Adicionar 3 unidades à coluna Price
- Utilizar **value_counts** para cada uma das colunas do dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales.csv')
```

```python
df.head(10)
```

```python
df["Account"] = df["Account"].apply(lambda x: x+1)
```

```python
list(df)
```

```python
df[" Price"] = df[" Price"].apply(lambda x: np.nan if x>50000 else x)
```

```python
df.head(10)
```

```python
pd.value_counts(df["Account"])
```

```python
df[" Price"] = df[" Price"].apply(lambda x: x+3)
```

```python
for c in df.columns:
    print ("---- %s ---" % c)
    print (df[c].value_counts().head())
```
